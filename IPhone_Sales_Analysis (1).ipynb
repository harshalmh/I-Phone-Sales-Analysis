#IPHONE SALES ANALYSIS

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv("apple_products.csv")
```

```python
data
```

```python
#to check whwether the data is correct or not or missing
print(data.isnull().sum())
```

```python
#to get statistical value
print(data.describe())
```

#I phone Sales In India

```python
import matplotlib.pyplot as plt

highest_rated = data.sort_values(by=["Star Rating"], ascending=False).head(10)


labels = highest_rated["Product Name"]
ratings = highest_rated["Star Rating"]

plt.figure(figsize=(10, 6))
plt.barh(labels, ratings, color='blue')
plt.xlabel("Star Rating")
plt.ylabel("iPhone Model")
plt.title("Top 10 Highest Rated iPhones")
plt.gca().invert_yaxis()

plt.show()
```

#Number of Highest Rating I phone On Flipkard

```python
import matplotlib.pyplot as plt


iphones = highest_rated["Product Name"].value_counts()

labels = iphones.index
counts = highest_rated["Number Of Ratings"]


plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='blue')


plt.title("No. of Ratings of Highest Rated iPhones")
plt.xlabel("iPhone Model")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))  # Adjust figure size
plt.scatter(data["Number Of Ratings"], data["Sale Price"], s=data["Discount Percentage"] * 5, alpha=0.5, color='blue')

plt.title("Relationship Between Sale Price and Number of Ratings")
plt.xlabel("Number of Ratings")
plt.ylabel("Sale Price")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))  # Set figure size
plt.scatter(data["Number Of Ratings"], data["Discount Percentage"], s=data["Sale Price"] * 0.1, alpha=0.5, color='blue')

plt.title("Relation Between Discount Percentage and Number of Ratings")
plt.xlabel("Number of Ratings")
plt.ylabel("Discount Percentage")
plt.grid(True)

plt.show()
```
